```python
import pandas as pd, numpy as np

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
total_school = school_data_complete["School ID"].nunique()
total_students = school_data_complete["Student ID"].count()
#Budget data should be taken from school data otherwise it will count all merged datas and gives wrong answer
total_budget = school_data["budget"].sum()
ave_mathscore = school_data_complete["math_score"].mean()
ave_readscore = school_data_complete["reading_score"].mean()
pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()
per_pass_math = pass_math/total_students*100
pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["reading_score"].count()
per_pass_read = pass_read/total_students*100
overall_pass = school_data_complete[(school_data_complete["math_score"]>= 70) 
                                    & (school_data_complete["reading_score"]>= 70)]["student_name"].count()
per_overall_pass = overall_pass/ total_students*100
#summary dataframe
district_summary = pd.DataFrame({"Total Schools":[total_school], "Total Students":[total_students], "Total Budget":[total_budget],
                                "Average Math Score":[ave_mathscore], "Average Reading Score":[ave_readscore], 
                                 "% Passing Math":[per_pass_math], "% Passing Reading":[per_pass_read], 
                                 "% Overall Passing":[per_overall_pass] })
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Groupby school_name from merged data for the rest calculations and hold bugdet and type datas from school dataframe
grouped_byschool = school_data_complete.set_index("school_name").groupby(["school_name"])
students_total = grouped_byschool["Student ID"].count()

per_stubudget = school_data.set_index("school_name")["budget"]/students_total
school_type = school_data.set_index("school_name")["type"]

mathscore_ave = grouped_byschool["math_score"].mean()
readscore_ave = grouped_byschool["reading_score"].mean()

math_pass = school_data_complete.loc[school_data_complete["math_score"] 
                                     >= 70].groupby("school_name")["Student ID"].count()/students_total*100        
read_pass = school_data_complete.loc[school_data_complete["reading_score"] 
                                     >= 70].groupby("school_name")["Student ID"].count()/students_total*100
overall = school_data_complete[(school_data_complete["math_score"] >= 70) 
                               & (school_data_complete["reading_score"] 
                                  >= 70)].groupby("school_name")["Student ID"].count()/students_total*100
#summary dataframe

#summary dataframe
school_summary = pd.DataFrame({ "School Type" : school_type, "Total Students" : students_total, "Total School Budget" : total_budget,
                               "Per Student Budget" : per_stubudget, "Average Math Score" : mathscore_ave,
                               "Average Reading Score" : readscore_ave, "% Passing Math": math_pass, "% Passing Reading" : read_pass,
                               "Overall Passing Rate" : overall })
#format
school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:.2f}".format)

school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 584, 629, 644, 675]
spending = ["<$584", "$585-629", "$630-644", "$645-675"]
school_data_complete["bins_spending"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels=spending)

grouped_bybins = school_data_complete.groupby("bins_spending")

math = grouped_bybins["math_score"].mean()
read = grouped_bybins["reading_score"].mean()
students_total = grouped_bybins["Student ID"].count()
passmath = school_data_complete.loc[school_data_complete["math_score"] 
                                     >= 70].groupby("bins_spending")["Student ID"].count()/students_total*100 
passread = school_data_complete.loc[school_data_complete["reading_score"] 
                                     >= 70].groupby("bins_spending")["Student ID"].count()/students_total*100 
passoverall = school_data_complete[(school_data_complete["math_score"] >= 70) 
                               & (school_data_complete["reading_score"] 
                                  >= 70)].groupby("bins_spending")["Student ID"].count()/students_total*100

spending_bybins = pd.DataFrame({"Average Math Score": math,
                                "Average Reading Score": read,
                                "% Passing Math":passmath,
                                "% Passing Reading":passread,
                             "% Overall Passing":passoverall })
                             
#format
spending_bybins["Average Math Score"]=spending_bybins["Average Math Score"].map("{:,.2f}".format)
spending_bybins["Average Reading Score"]=spending_bybins["Average Reading Score"].map("{:.2f}".format)
spending_bybins["% Passing Math"]=spending_bybins["% Passing Math"].map("{:,.2f}".format)
spending_bybins["% Passing Reading"]=spending_bybins["% Passing Reading"].map("{:,.2f}".format)
spending_bybins["% Overall Passing"]=spending_bybins["% Overall Passing"].map("{:,.2f}".format)
spending_bybins.index.name = "Spending Ranges (Per Student)"
spending_bybins 
```
